| Feature | Flat        | IVF (Inverted File Index)         | HNSW (Graph-based Index)         |
|---------|-------------|------------------------------------|----------------------------------|
| Type of Search | Exact        | Approximate (cluster-based)     | Approximate (graph-based traversal) |
| Speed   | Slow (linear scan) | Fast (search only in top clusters) | Very Fast (graph walk)           |


| Dataset Size | Recommended Index                             |
|--------------|------------------------------------------------|
| UPTO 1L      | IndexFlatL2 or IndexFlatIP                    |
| UPTO 1M      | IndexIVFFlat or IndexHNSWFlat                 |
| > 1M         | IndexIVFPQ or IndexHNSWFlat                   |


```python
# from uuid import uuid4
from langchain_core.documents import Document

document_1 = Document(
    page_content="I had chocolate chip pancakes and scrambled eggs for breakfast this morning.",
    metadata={"source": "tweet"},
)

document_2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
)

document_3 = Document(
    page_content="Building an exciting new project with LangChain - come check it out!",
    metadata={"source": "tweet"},
)

document_4 = Document(
    page_content="Robbers broke into the city bank and stole $1 million in cash.",
    metadata={"source": "news"},
)

document_5 = Document(
    page_content="Wow! That was an amazing movie. I can't wait to see it again.",
    metadata={"source": "tweet"},
)

document_6 = Document(
    page_content="Is the new iPhone worth the price? Read this review to find out.",
    metadata={"source": "website"},
)

document_7 = Document(
    page_content="The top 10 soccer players in the world right now.",
    metadata={"source": "website"},
)

document_8 = Document(
    page_content="LangGraph is the best framework for building stateful, agentic applications!",
    metadata={"source": "tweet"},
)

document_9 = Document(
    page_content="The stock market is down 500 points today due to fears of a recession.",
    metadata={"source": "news"},
)

document_10 = Document(
    page_content="I have a bad feeling I am going to get deleted :(",
    metadata={"source": "tweet"},
)


#Creating the List Of the Documents
documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]
```

#### This creates a FAISS index using:

- `IndexFlatIP:` a flat (non-approximate) index using Inner Product (IP) as the similarity metric.

- `384:` the dimension of the vectors (e.g., OpenAI text-embedding-ada-002 produces 1536-dim, some others 384-dim)

```python
#Creating the Index
index = faiss.IndexFlatIP(384)

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
vector_store.add_documents(documents = documents)
```

```python
vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
   # k=2 #hyperparameter
    
)
```

```python
vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
   k=2 #hyperparameter
    
)
```

```python
vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
    #k=2 #hyperparameter,
    filter={"source":{"$eq": "tweet"}}
    
)
```

```python
result = vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
    #k=2 #hyperparameter,
    filter={"source":"news"}
    
)
```

```python
result[0].metadata
```

```python
## It will able to load faster
pages = []
async for page in loader.alazy_load():
    pages.append(page)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,#hyperparameter
    chunk_overlap=50 #hyperparemeter
)
```

```python
split_docs = splitter.split_documents(pages)
```

```python
len(split_docs)
```

```python
index=faiss.IndexFlatIP(384)
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
vector_store.add_documents(documents=split_docs)
```

```python
retriever=vector_store.as_retriever(
    search_kwargs={"k": 10} #hyperparameter
)
```

```python
retriever.invoke("what is llama model?")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
# context(retriever),
# prompt(hub),
# model(google),
# parser(langchain)
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
    
```

```python
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```

```python
rag_chain.invoke("what is llama model?")
```
